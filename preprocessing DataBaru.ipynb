```python
import csv
import string
import Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.tokenize import word_tokenize
import pandas as pd
```

```python
filedataolah = 'DataBaru.csv'
f = open(filedataolah,'r')
reader = csv.reader(f, delimiter=';')
```

```python
def casefolding(judul):
    judul = judul.lower()
    return(judul)
```

```python
hasil_case_folding = []
for data in reader:
    hasil = casefolding(data[0])
    hasil_case_folding.append(hasil)
kolom = hasil_case_folding.pop(0)
print(hasil_case_folding)   
```

```python
print(hasil_case_folding)
```

```python
def remove_punctuation(judul):
    judul = judul.translate(str.maketrans('','',string.punctuation)).strip()
    return(judul)
```

```python
hasil_remove_punctuation = []
for data in range(len(hasil_case_folding)):    
    hasil = remove_punctuation(hasil_case_folding[data])
    hasil_remove_punctuation.append(hasil)
print(hasil_remove_punctuation)
```

```python
print (hasil_remove_punctuation[1])
```

```python
def tokenize(judul):
    judul = word_tokenize(judul)
    return(judul)
```

```python
hasil_tokenize = []
for data in range(len(hasil_remove_punctuation)):    
    hasil = tokenize(hasil_remove_punctuation[data])
    hasil_tokenize.append(hasil)
print(hasil_tokenize)
```

```python
print(hasil_tokenize[0])
```

```python
from nltk.corpus import stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
def stopwords(judul): 
    
    hasil_judul = []
    for i in judul:
        word = stopword.remove(i)
        if word !='':
            hasil_judul.append(word)
    return(hasil_judul) 
```

```python
hasil_stopwords = []
for data in range(len(hasil_tokenize)):    
    hasil = stopwords(hasil_tokenize[data]) 
    hasil_stopwords.append(hasil)
print(hasil_stopwords)
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer() 
def stemming(judul):  
    
    hasil_judul = []
    for i in judul:
        word = stemmer.stem(i)
        if word !='':
            hasil_judul.append(word)
    return(hasil_judul)
```

```python
hasil_stemming = []  
for data in range(len(hasil_stopwords)):    
    hasil = stemming (hasil_stopwords[data]) 
    hasil_stemming.append(hasil) 
print(hasil_stemming)
```

```python
hasil_preprocessingDataBaru = hasil_stemming
json_preprocessing = {
    'casefolding':hasil_case_folding,
    'remove_punctuation':hasil_remove_punctuation,
    'tokenizing':hasil_tokenize,
    'stopwords':hasil_stopwords,
    'hasil_preprocessing':hasil_preprocessingDataBaru
}
```
